##Why should I be interested in survival analysis?

- I'm going to talk about why a data analyst or data scientist might need survival analysis, from a high point of view, and in the next screencasts we'll start to get more hands on. 

- Historically, survival analysis was used as a way to measure lifetimes of populations. For example, an actuary would like to measure a nations population, eg., "what is the expected lifespan of a newborn born today?". Health professionals use survival analysis to measure lifespans of individuals afflicted with disease, or to show effectiveness of some drug. 

- They do this by first defining a "birth" and "death" event. Now keep in mind, a "birth" event is not always a birth, and a "death" is not always a death. These are very general terms in survival analysis. For example, a "birth" could be an individual being diagnosed with a disease, and a "death" even could be their death, OR it could be their recovery, depending on what duration you are interested in measure. 

- Another example: the birth event could be a new webuser signing up for a online subscription service, and the "death" event is when they leave. Furthermore, survival analysis provides the solution to the lifetime value problem, that is, how much value does this customer provide me, which we'll get to in another screencast. These reasons are why telcoms and SaaS business are so interested in survival analysis right now. 

- So where does the difficulty lie? We'll let's look at some data: 


- Suppose at time 0, 25 user join our service. At time 10, I want to measure the average lifetime of a user on my service. But at time 10, only 21 individuals have left my service - 4 of them are still on.

```python

observed_lifetimes = np.array([ 3,  7,  6,  1,  3,  
                                3,  6,  0,  6,  7, 
                                10, 10,  9,  3,  2,  
                                2,  7, 4, 10,  2, 10,  
                                0,  1,  3,  1])
observed = array([ True,  True,  True,  True,  True,  True,  True,  True,  True,
                    True, False, False,  True,  True,  True,  True,  True,  True,
                   False,  True, False,  True,  True,  True,  True], dtype=bool)
```

```python
%pylab inline
figsize(12,8)

from lifelines.plotting import plot_lifetimes

plt.xlim(0,25)
plt.vlines(10,0,30,lw=2, linestyles="--")
plt.xlabel('time')
plt.title('Births and deaths of our population, at $t=10$')
plot_lifetimes(observed_lifetimes, event_observed=observed)
```

So how do I measure the average lifetime?

1. Only use the individuals who died, or left the service:

```python
observed_lifetimes[observed].mean()
```

2\. But that's not really fair: the customers who currently exist have survived much longer, and should be included in the estimate. Okay, let's include them, and use their observed lifeltime so far:

```python
observed_lifetimes.mean()
```

Clearly we were underestimating the average by excluding the users who still exist. But is the above number correct? The current users could live for many more months, or they could leave tomorrow. The problem is, I haven't seen their death event. What's the correct value? In fact, let's peak into the future and look at their actual lifespans:

```python
actual_lifetimes = np.array([ 3,  7,  6,  1,  3,  3,  6,  0,  6,  7, 24, 55,  9,  3,  2,  2,  7,
        4, 12,  2, 33,  0,  1,  3,  1])
plt.xlim(0,25)
plt.vlines(10,0,30,lw=2, linestyles="--")
plt.xlabel('time')
plt.title('Births and deaths of our population, at $t=10$')
plot_lifetimes(actual_lifetimes, event_observed=observed)
```

Clearly our estimate of 4.6 months is considerably off. What to do?


### Censorship

The lack of complete information is called censorship in survival analysis, and it's existance is the reason for the development and use of survival analysis. Consider: if everyone had left, then it would be easy to measure the expected lifetime was: just take the mean. But if we haven't seen all the deaths, because they exist in the future or another factor has hidden them, then we say that individual has been censored. 

More specifically, they have been right censored. In the above example, the blue individual's deaths have been right censored. Mathematically, If we denote the current time as small $t$, and an individual lifetime as big $T$, then they are censored if 

$$  t < T $$

that is big T, their lifetime,occurs after the current time, small $t$. 

### Where else does censorship creep in?

- For actuaries, they are interested in measuring a life expectancy of a population, but not everyone in that population has died yet. 
- For health professionals interested in drug trials, they do not want to wait until everyone has died before looking at the effectiveness of a drug.
- For telcoms, they want to acurately measure how long customers stay on their service for. 
- A sociologist is interested in measure how long it takes couples to divorce, a "birth" is when the couple marries, and "death" is when they divorce, but not every couple divorces: so censorship occurs when the first partner dies (as you can't divorce after that). This is a case where a T, the time of "death", can be infinity, and that's okay. 
- As users enter your service, you might be interested in how long it takes them to complete a task. Individuals who have not completed the task are right-censored, and may never complete the task, or will, but much farther in the future. 
- How long does it take the FBI to catch a criminal? Well you could take the average of the duration it took to capture all *captured criminals*, but this is ignores the criminals who were never captured!
